```python
import json

with open('db/books_geometry.json', 'r') as f:
    books = json.load(f)
```

```python
len(books)
```

```python
import pandas as pd

df = pd.DataFrame.from_records(books).dropna(subset=['cover'])[['authors', 'book_id', 'cover', 'subject', 'theme', 'name']]
df.cover = df.cover.apply(lambda x: x['path'])
```

```python
df.to_csv('books_geo.csv', sep=',', index=False)
df.head()
```

```python
df_2 = pd.DataFrame.from_records(books).dropna(subset=['cover'])[['authors', 'book_id', 'cover', 'subject', 'theme', 'name']]
df_2.cover = df_2.cover.apply(lambda x: x['path'])
```

```python
df_3 = pd.concat([df, df_2], ignore_index=True)
df_3.head()
```

```python
len(df_3)
```

```python
df_3 = df_3.sort_values(by='name').drop_duplicates(subset=['cover'])
df_3.to_csv('books.csv', sep=',', index=False)
```

```python
import json 
import pandas as pd

df_1 = pd.read_csv('books_masha.tsv', sep='\t').drop_duplicates(subset=['book_id'])
# df_1.columns = ['authors', 'book_id', 'cover', 'subject', 'theme', 'name', 'kes1', 'kes2', 'kes3']
df_1['kes3'] = df_1['kes3'].apply(lambda x: x+100)
df_2 = pd.read_csv('books_masha_2.tsv', sep='\t')
df = pd.concat([df_1, df_2], ignore_index=True)
df['kes'] = df[['kes1', 'kes2', 'kes3']].apply(lambda row: f"{row.kes1}.{row.kes2}.{row.kes3}", axis=1)
df.to_csv('db/materials.csv', index=False)
df.head()
```

```python
import json 

mapping = {}

for i in range(len(df)):
    key = df.iloc[i]['subject'].split(',')[0][7:]
    if not mapping.get(key):
        mapping[key] = df.iloc[i]['subject'].split(',')[1][10:-2]
        
print(mapping)

with open('db/mapping.json', 'w') as f:
    json.dump(mapping, f, indent=4)
```
